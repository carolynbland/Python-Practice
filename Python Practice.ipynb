# <span style="color:blue">Topic 1: Statistics </span>

### <span style="color:green">Basic Definitions </span>

* <font size="2">**Statistics** = Collecting, organizing, and interpreting data, as well as drawing conclusions and making decisions</font>
* <font size="2">**Variable/Feature** = Any property measured/observed in a study</font>
> <font size="2">E.g. porosity, permeability, mineral concentrations, saturations, contaminant concentration </font>
> <font size="2">Measure often requires significant analysis, interpretation etc. </font>
* <font size="2">**Population** = Exhaustive, finite list of property of interest over area of interest. Generally, the entire population is not accessible</font>
> <font size="2">E.g. exhaustive set of porosity at each location within a reservoir </font>
* <font size="2">**Sample**= The set of values, locations that have been measured </font>
> <font size="2">A finite set of measurements over an area of interest </font>
> <font size="2">E.g. porosity data from well-logs within a reservoir</font>
* <font size="2">**Parameters**= Summary measure of a population </font>
> <font size="2">E.g. population mean, population standard deviation, we rarely have access to this </font>
* <font size="2">**Statistic**= Summary measure of a sample </font>
> <font size="2">E.g. sample mean, sample standard deviation, we use statistics as estimates of the parameters  </font>


### <span style="color:green">Comparisons </span>
* **Population vs Sample:** A population is the total property of interest over the area of interest whereas a sample is a set of values and locations that have been measured 
* **Qualitative vs Quantitative:** qualitative is a feature that requires interpretation whereas quantitative is a feature you can measure directly 
* **Estimation vs Simulation:** An estimation is a statistical model for local accuracy whereas a simulation is a statistical model for global accuracy 
* **Prediction vs Inference:** Prediction is predicting the sample from a given model of the population whereas an inference is describing the population given a sample

### <span style="color:green">Types of Statistics </span>
* **Inferential Statistics:** given a random sample from a population, describe the population
> E.g. given the well(s) samples, describe the reservoir 

* **Predictive Statistics:** predict the samples given assumptions about the population 
> Given our model of the reservoir, predict the next well (pre-drill assessment) sample, e.g. porosity, permeability, production rate etc


### <span style="color:green">Basic Definitions </span>
* Deterministic Model: Assumes the system or process that is completely predictable 
> Engineering and geoscience physics and expert judgement 
> E.g. interpretation of 3D seismic data, physics-based flow modeling 
> Advantages: Integration of physics and expert knowledge, integration of various information sources 
> Disadvantages: often quite time consuming, often no assessment of uncertainty, focus on building one model 
* Statistical Model: system or process that is uncertain, multiple models constrained by statistics 
> Data-driven, machine learning 
> E.g. Data-drive, geostatistics, machine learning 
> Advantages: speed, uncertainty assessment, report significance and confidence intervals, honor many types of data, data-driven approaches 
> Disadvantages: limited physics used, statistical model assumptions/simplification 

#### Types of Statistical Models:
* Estimation
> Process of obtaining the single best value of a reservoir property at an unsampled location 
> Local accuracy takes precedence over global spatial variability 
> Too ‘smooth’, not appropriate for forecasting 
> Most of machine learning! 
* Simulation (preferred) 
> Process of obtaining one or more good values of a reservoir property at an unsampled location 
> Global accuracy, matches the global statistics 
> Simulation methods tend to produce more realistic property distributions 
> Preferred because we can capture full distributions of properties of interest, the extremes matter and we need realistic models for flow simulation 

Transfer Function: Calculation applied to the model (deterministic or statistical) to calculate a decision criterion (maps 3D subsurface model to a decision criteria, brings data to something you can make a decision on)
-	E.g. volumetric calculation for oil-in-place, flow simulation for production forecast 
Uncertainty Modeling 
-	Measurement of amount of error 
-	Generate multiple simulated models that jointly represent a space of uncertainty 
-	Realizations: hold input parameters constant and change random number seed 
-	Scenarios: change the input parameters or other modeling choices 
-	Reservoir modeling uncertainty process 1) integrate all available information and build multiple scenarios and realizations to sample the uncertainty space 2) apply all models to transfer function to map a decision criteria 3) Assemble the distribution of the decision criteria 4) Make the best decision accounting for this uncertainty model 

#### Types of Data	
* Hard data: the data that has a high degree of certainty, usually based on a direct measurement 
> E.g. well core and log based porosity, lithofacies assessed from well logs 
* Soft data: data that provides indirect measures of the property of interest with a significant degree of uncertainty 
> E.g. probability density function for local porosity calibrated from acoustic impedance 
* Primary data: the variable of interest and target for building a model 
> E.g. porosity measures from cores and logs used to build a full 3D porosity model 
* Secondary Data: another variable/feature that provides information about the primary data through a relationship/calibration 
> E.g. acoustic impedance to support modeling porosity and porosity to support modeling permeability 

#### Describing Data	
* Coverage: Volume of covered by all the samples of this type of data what proportion of the reservoir/population has this data available typically? 
> E.g. a couple meters around wells 
* Scale/Support Size: scale of the individual sample measurements. what is the scale of the individual data measures? 
> E.g. pore scale 
* Information Type: type of information from this type of data. what does the data tell us about the subsurface? 
> E.g. grain size, fluid type 
* Resolution: ratio between the maximum signal measured to the smallest part that can be resolved or smallest change that can be measured (the scale of the individual sample measurements) 

#### Basic nomenclature	
* Sample 
> Proportion (p̂)
> Mean (x̅)
> Standard deviation (s)
> Variance (s2)

* Population 
> Proportion (p)
> Mean (µ)
> Standard deviation (σ)
> Variance (σ2)
